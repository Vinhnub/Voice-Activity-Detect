```python
# time domain
import librosa
import numpy as np
import matplotlib.pyplot as plt

DIR = r"C:\Users\admin\Documents\PythonFile\Project\Voice-Activity-Detect\data\processed\speech\mixed\1.wav"
DIR = DIR.replace('\\', '/')
y, sr = librosa.load(DIR, sr=None)

t = np.arange(len(y)) / sr

plt.figure(figsize=(10, 4))
plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform of Audio Signal")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# hz - domain

import librosa
import numpy as np
import matplotlib.pyplot as plt

DIR = r"C:\Users\admin\Documents\PythonFile\Project\Voice-Activity-Detect\data\processed\speech\mixed\1.wav"
DIR = DIR.replace('\\', '/')
y, sr = librosa.load(DIR, sr=None)

N = len(y)
Y = np.fft.rfft(y)
freqs = np.fft.rfftfreq(N, d=1/sr)

mag = np.abs(Y)

# Plot frequency domain
plt.figure(figsize=(10, 4))
plt.plot(freqs, mag)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency Domain (FFT)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
