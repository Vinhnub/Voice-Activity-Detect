```python
import librosa
import numpy as np
import os

def extract_mfcc(y, sr, n_mfcc=39, n_fft=400, hop_length=160, n_mels=40):
    mfcc = librosa.feature.mfcc(
        y=y,
        sr=sr,
        n_mfcc=n_mfcc,
        n_fft=n_fft,
        hop_length=hop_length,
        n_mels=n_mels
    )

    # CMVN
    return mfcc.T   # (num_frames, n_mfcc)
```

```python
def build_mfcc_dataset(
    wav_list,
    sr=16000,
    out_path="X_mfcc.npy"
):
    all_feats = []

    for wav_path in wav_list:
        print(wav_path)
        y, sr = librosa.load(wav_path, sr=sr)
        mfcc = extract_mfcc(y, sr)
        all_feats.append(mfcc)

        print(f"{wav_path} -> {mfcc.shape}")

    # Gộp theo frame
    X = np.vstack(all_feats)
    #X = X[:int(X.shape[0]//10), :]
    np.save(out_path, X)
    print("Saved:", out_path, X.shape)

    return X
```

```python
DIR = "E:/PythonFile/Project/Voice-Activity-Detect/data/processed/aurora/speech-mixed"

wav_list = [f"{DIR}/{i + 1}.wav" for i in range(180)]
#wav_list.pop(766)

X_mfcc = build_mfcc_dataset(wav_list, out_path="E:/PythonFile/Project/Voice-Activity-Detect/data/feature/train_mini/aurora_mfcc_speech_mixed.npy")
print(X_mfcc.shape)
```
