```python
import librosa
import numpy as np
import os

def extract_mfcc(y, sr, n_mfcc=13, n_fft=400, hop_length=160, n_mels=40):
    mfcc = librosa.feature.mfcc(
        y=y,
        sr=sr,
        n_mfcc=n_mfcc,
        n_fft=n_fft,
        hop_length=hop_length,
        n_mels=n_mels
    )

    # CMVN
    return mfcc.T   # (num_frames, n_mfcc)
```
