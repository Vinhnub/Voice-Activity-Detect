```python
import numpy as np

def chroma_from_stft(stft_matrix, sr, n_fft):
    """
    stft_matrix: shape (freq_bins, time_frames)
    sr          : sample rate
    n_fft       : FFT size
    """
    num_bins, num_frames = stft_matrix.shape
    chroma = np.zeros((12, num_frames))

    for t in range(num_frames):
        for k in range(1, num_bins):
            magnitude = stft_matrix[k, t]
            f_k = k * sr / n_fft

            if f_k < 20:
                continue

            p = 12 * np.log2(f_k / 440.0)
            chroma_idx = int(np.round(p)) % 12
            chroma[chroma_idx, t] += magnitude

    return chroma
```

```python
import os
import numpy as np

stft_dir   = r"D:\FPT_prj\preprocess_data\speech\stft"
chroma_dir = r"D:\FPT_prj\preprocess_data\speech\chroma"
os.makedirs(chroma_dir, exist_ok=True)

sr = 16000
n_fft = 512

for i in range(27, 101):
    stft_path = os.path.join(stft_dir, f"{i}_stft.npy")

    if not os.path.exists(stft_path):
        print(f"Missing STFT: {i}_stft.npy")
        continue

    # load STFT
    stft = np.load(stft_path)    # (time, freq)

    # transpose → (freq, time)
    stft = stft.T

    # chroma
    chroma = chroma_from_stft(stft, sr, n_fft)

    # log-compression (optional but recommended)
    chroma = np.log(chroma + 1e-8)

    # CMVN (optional)
    chroma = (chroma - chroma.mean(axis=1, keepdims=True)) / \
             (chroma.std(axis=1, keepdims=True) + 1e-8)

    # save
    save_path = os.path.join(chroma_dir, f"{i}_chroma.npy")
    np.save(save_path, chroma)

    print(f"{i}_stft.npy → chroma shape: {chroma.shape}")
```
