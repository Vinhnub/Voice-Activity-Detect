```python
import os
import librosa
import numpy as np

def build_stft_dataset_librosa(
    wav_dir,
    sr=16000,
    frame_len=400,
    hop_len=160,
    n_fft=512,
    out_path="X_stft_librosa.npy"
):
    all_specs = []

    wav_files = sorted([
        os.path.join(wav_dir, f)
        for f in os.listdir(wav_dir)
        if f.lower().endswith(".wav")
    ])
    for wav_path in wav_files:
        print(wav_path)
        signal, _ = librosa.load(wav_path, sr=sr, mono=True)

        S = librosa.stft(
            signal,
            n_fft=n_fft,
            hop_length=hop_len,
            win_length=frame_len,
            window="hann",
            center=False
        )

        spec = np.abs(S).T
        all_specs.append(spec)

    all_specs = np.vstack(all_specs)
    np.save(out_path, all_specs)
    print(all_specs.shape)
    return all_specs
```

```python
DIR = "D:/Pythonfile/Voice-Activity-Detect/data/processed/speech/non-mixed" 
OUT = "D:/Pythonfile/Voice-Activity-Detect/data/feature/train"
X_stft = build_stft_dataset_librosa(
    wav_dir=DIR,
    frame_len=400,
    hop_len=160,
    n_fft=512,
    out_path=f"{OUT}/stft_speech.npy"
)
```
