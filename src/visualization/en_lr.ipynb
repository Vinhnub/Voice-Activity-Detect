```python
import numpy as np
import matplotlib.pyplot as plt

# Feature names
features = [
    "MFCC",
    "MFCC + Δ",
    "MFCC + Δ + ΔΔ",
    "MFCC + ΔΔ",
    "LFCC",
    "LFCC + Δ",
    "LFCC + Δ + ΔΔ",
    "LFCC + ΔΔ",
    "STFT"
]

# Metrics data (theo bảng bạn đưa)
accuracy  = [0.848, 0.848, 0.852, 0.843, 0.790, 0.824, 0.824, 0.824, 0.824]
precision = [0.838, 0.838, 0.842, 0.833, 0.787, 0.821, 0.821, 0.821, 0.798]
recall    = [0.866, 0.866, 0.869, 0.862, 0.792, 0.827, 0.827, 0.827, 0.872]
f1_score  = [0.852, 0.852, 0.855, 0.847, 0.790, 0.824, 0.824, 0.824, 0.833]

x = np.arange(len(features))
width = 0.2

plt.figure(figsize=(14, 6))

plt.bar(x - 1.5*width, accuracy,  width, label="Accuracy")
plt.bar(x - 0.5*width, precision, width, label="Precision")
plt.bar(x + 0.5*width, recall,    width, label="Recall")
plt.bar(x + 1.5*width, f1_score,  width, label="F1-score")

plt.xticks(x, features, rotation=30, ha="right")
plt.ylabel("Score")
plt.ylim(0.75, 0.90)
plt.title("Performance Comparison of Audio Features")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
