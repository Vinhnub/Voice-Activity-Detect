```python
import numpy as np
import matplotlib.pyplot as plt

# Feature names
features = [
    "MFCC",
    "MFCC + Δ",
    "MFCC + Δ + ΔΔ",
    "MFCC + ΔΔ",
    "LFCC",
    "LFCC + Δ",
    "LFCC + Δ + ΔΔ",
    "LFCC + ΔΔ",
    "STFT"
]

# Metrics (theo bảng mới)
accuracy  = [0.894, 0.904, 0.906, 0.896, 0.839, 0.839, 0.871, 0.841, 0.839]
precision = [0.878, 0.890, 0.892, 0.879, 0.829, 0.830, 0.860, 0.831, 0.829]
recall    = [0.918, 0.922, 0.924, 0.920, 0.853, 0.854, 0.888, 0.857, 0.853]
f1_score  = [0.897, 0.906, 0.908, 0.899, 0.841, 0.841, 0.873, 0.844, 0.841]

x = np.arange(len(features))
width = 0.2

plt.figure(figsize=(14, 6))

plt.bar(x - 1.5*width, accuracy,  width, label="Accuracy")
plt.bar(x - 0.5*width, precision, width, label="Precision")
plt.bar(x + 0.5*width, recall,    width, label="Recall")
plt.bar(x + 1.5*width, f1_score,  width, label="F1-score")

plt.xticks(x, features, rotation=30, ha="right")
plt.ylabel("Score")
plt.ylim(0.80, 0.95)
plt.title("Performance Comparison of Audio Features")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
