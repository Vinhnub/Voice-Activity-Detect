```python
import numpy as np

from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, cross_validate
```

```python
X_1 = np.load("C:/Users/Vinh/Documents/Pythonfile/Voice-Activity-Detect/data/feature/test/speech/mfcc_speech.npy")  # label = 1
X_0 = np.load("C:/Users/Vinh/Documents/Pythonfile/Voice-Activity-Detect/data/feature/test/non-speech/mfcc_non-speech.npy")  # label = 0

print("X_1:", X_1.shape)
print("X_0:", X_0.shape)
print(X_1.shape, X_0.shape)
y_1 = np.ones(X_1.shape[0], dtype=int)   # toàn 1
y_0 = np.zeros(X_0.shape[0], dtype=int)  # toàn 0
X = np.vstack([X_1, X_0])
y = np.hstack([y_1, y_0])

print("X:", X.shape)
print("y:", y.shape)
idx = np.random.permutation(len(y))
X = X[idx]
y = y[idx]
print(y)
```

```python
import matplotlib.pyplot as plt

plt.hist(X_1[:, 0], bins=100, alpha=0.5, label="speech")
plt.hist(X_0[:, 0], bins=100, alpha=0.5, label="non-speech")
plt.legend()
plt.show()
```

```python
svm_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", LinearSVC(
        C=1.0,          # soft-margin
        loss="hinge",
        max_iter=10000
    ))
])
```

```python
cv = StratifiedKFold(
    n_splits=5,
)
```

```python
scoring = {
    "acc": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1"
}

cv_results = cross_validate(
    svm_model,
    X,
    y,
    cv=cv,
    scoring=scoring,
    n_jobs=-1
)
```

```python
for metric in scoring.keys():
    scores = cv_results[f"test_{metric}"]
    print(f"{metric.upper():9s} | "
          f"Mean = {scores.mean():.4f} | "
          f"Std = {scores.std():.4f}")
```
